**3. Transfer Learning Model**

> **Preprocessing data**

>Menggunakan teknik **RobustScaler** untuk prapemrosesan di scikit-learn yang menskalakan fitur numerik dengan menghapus median dan menskalakan data berdasarkan Rentang Interkuartil (persentil ke-25 hingga ke-75). dan sangat efektif untuk dataset dengan outlier yang signifikan, sehingga memastikan outlier tersebut tidak secara berlebihan memengaruhi distribusi fitur.

```python
from sklearn.preprocessing import RobustScaler

numeric_cols = ['Harga', 'Usia_Bangunan', 'Furnish', 'Kota', 'Kecamatan', 'Lalitude', 'Longitude', 'Karpot', 'Garasi', 'Kamar Tidur']

rs = RobustScaler()

data[numeric_cols] = rs.fit_transform(data[numeric_cols])

print("Setelah dilakukan normalisasi:")

print(data[numeric_cols].head())
```

>Analisis Komponen Utama (PCA) adalah teknik reduksi dimensi dari scikit-learn yang digunakan untuk mengubah dataset berdimensi tinggi menjadi himpunan variabel yang lebih kecil dan tidak berkorelasi (komponen utama) sambil mempertahankan sebanyak mungkin varian.

```python
from sklearn.decomposition import PCA

df_scaled = data[numeric_cols]
df_scaled
```

```python
X_scaled = df_scaled.values #ubah kedalam bentuk array
X_scaled
```

```python
pca = PCA(n_components=6) #set ke-6 jumlah komponen pada PCA

X_pca = pca.fit_transform(X_scaled)
X_pca
```

>Algoritma K-Means digunakan untuk mengelompokkan data (clustering) tanpa label (unsupervised learning) ke dalam jumlah klaster berdasarkan kemiripan karakteristik atau fitur data, dengan tujuan meminimalkan jarak antara titik data dan pusat klaster (centroid).

>WCSS digunakan untuk menentukan klaster optimal berdasarkan kepadatan dan kualitas klaster serta jumlah kuadran jarak di setiap klaster dan pusat-Nya.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
K_range = range(1,11)
wcss = []  #with cluster sum of square

# Dilakukan perulangan untuk mencari klaster optimal dari k_range
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_pca)
    wcss.append(kmeans.inertia_)
```

>I**ndikator** : Semakin kecil skor WCSS, semakin baik. Sumbu x adalah jumlah kluster, sumbu y adalah skor WCSS.

>**Elbow Method** adalah teknik heuristik yang digunakan dalam analisis kluster untuk menentukan jumlah kluster optimal
 pada dataset, khususnya untuk algoritma K-Means.

```python
plt.figure(figsize=(10,6))
plt.plot(K_range, wcss, marker='o')
plt.title("Grafik elbow untuk menentukan jumlah cluster yang optimal")
plt.xlabel('Jumlah cluster')
plt.ylabel('with cluster sum of square (WCSS)')
plt.grid(True)
plt.show()
```

>Bisa dilihat bahwa saat K=1, nilai WCSS sangat tinggi. Kemudian menurun terus sampai K=10 terlihat membentuk seperti sebuah siku. Mulai K=4 sampai K=10 penurunan skor WCSS sudah tidak signifikan. Dengan demikian, dapat diketahui bahwa jumlah kluster yang tepat untuk grafik di atas adalah 3.

```python
# Inisiasi klaster untuk menampung nilai klaster yaitu 3 dan Inisiasi y sebagai nilai prediksi dari X_scaled
klaster = KMeans(n_clusters=3, random_state=42)
y = klaster.fit_predict(X_scaled)
```

>Visualisasi klaster dengan PCA berdasarkan 3 kategori klaster Rumah, yaitu **Menengah, Premium dan Premium Atas**

```python
plt.figure(figsize=(14, 9))

plt.scatter(X_pca[y == 0, 0], X_pca[y == 0, 1],
            marker='o', label='Cluster 0 - MENENGAH')

plt.scatter(X_pca[y == 1, 0], X_pca[y == 1, 1],
            marker='s', label='Cluster 1 - PREMIUM')

plt.scatter(X_pca[y == 2, 0], X_pca[y == 2, 1],
            marker='^', label='Cluster 2 - PREMIUM ATAS')

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Visualisasi Hasil K-Means (K=3) Menggunakan PCA")

# Centroid adalah titik tengah cluster dengan nilai dan digunakan untuk menghitung jarak suatu obyek data terhadap centroid
centroids_pca = pca.transform(klaster.cluster_centers_)

plt.scatter(centroids_pca[:, 0],
            centroids_pca[:, 1],
            marker='X',
            s=50,
            label='Centroid')

plt.legend()
plt.grid(True)
plt.show()
```

>Berdasarkan visualisasi hasil K-Means (K=3) menggunakan PCA, model berhasil mengelompokkan data ke dalam tiga segmen yang cukup terpisah dengan baik, yaitu Menengah, Premium, dan Premium Atas. Pemisahan cluster terlihat cukup jelas terutama pada komponen utama pertama (PC1), yang menjadi pembeda dominan antar kelompok. Cluster Menengah memiliki sebaran data yang relatif rapat dan terpusat pada nilai komponen yang lebih rendah, menunjukkan karakteristik yang homogen dan tidak ekstrem. Cluster Premium berada di posisi tengah dengan nilai komponen yang lebih tinggi dibandingkan Menengah, mencerminkan peningkatan karakteristik fitur. Sementara itu, Cluster Premium Atas memiliki nilai komponen paling tinggi dengan variasi yang lebih luas serta beberapa outlier, yang mengindikasikan segmen dengan karakteristik paling dominan dan heterogen. Secara keseluruhan, metode K-Means dengan tiga cluster mampu mengidentifikasi segmentasi data secara efektif dan memberikan gambaran yang jelas mengenai perbedaan tingkat karakteristik antar kelompok.

> **Splitting data**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,
    test_size=0.2, # Training 80% dan 20% untuk testing
    random_state=42
    )
```

```python
from sklearn.linear_model import LogisticRegression

# Inisiasi model logistik regresi yang diigunakan untuk menkategorikan klaster
logreg = LogisticRegression(
    solver='lbfgs',
    max_iter=100
)

logreg.fit(X_train, y_train)
```

**4. Evaluasi Model**

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import numpy as np

y_pred = logreg.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
y_pred = logreg.predict(X_test)

cm = confusion_matrix(y_test, y_pred)

plt.figure()
plt.imshow(cm)
plt.title("Confusion Matrix - Logistic Regression")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.colorbar()

classes = np.unique(y)

plt.xticks(range(len(classes)), classes)
plt.yticks(range(len(classes)), classes)

for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], ha='center', va='center')

plt.show()
```

>Berdasarkan confusion matrix Logistic Regression tersebut, model menunjukkan kinerja yang sangat baik dalam mengklasifikasikan data. Dari total 711 data (686 kelas 0 dan 25 kelas 1), model berhasil memprediksi 685 data kelas 0 dengan benar dan 25 data kelas 1 dengan benar. Hanya terdapat 1 kesalahan prediksi, yaitu satu data yang sebenarnya kelas 0 namun diprediksi sebagai kelas 1, dan tidak ada kesalahan untuk kelas 1 yang diprediksi sebagai kelas 0. Hal ini menunjukkan bahwa model memiliki tingkat akurasi yang sangat tinggi serta kemampuan yang sangat baik dalam mengenali kedua kelas, terutama karena tidak ada kasus kelas 1 yang terlewat. Secara keseluruhan, model Logistic Regression dapat dikatakan sangat efektif dan andal dalam melakukan klasifikasi pada data ini.
